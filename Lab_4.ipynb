```python
def CreateDict(data):
    output = {}
    for line in data.split("\n"):
        if line and not line[0] == " ":
            info = line.split(" ", 1)
            if len(info) > 1:
                output[info[0]] = [info[1].lstrip(" ")]
        else:
            output[list(output.keys())[-1]].append(line.lstrip(" "))
    return output
```
